```python
#Code for the merge sort algorithm solved recursively. 


#For the original merge sort program, I take in the array itself, the leftmost index, and the rightmost index. 

#Merges two subarrays
#First subarray is arr[l, m]
#Second subarray is arr[m+1, r]
def merge(arr, l, m, r):
    #We are going to create an empty array to merge the lists. 
    len_tempLM = m - l + 1
    #Don't have to add +1 because starts at m+1
    len_tempMR = r -m

    tempLM = [0] * len_tempLM
    tempMR = [0] * len_tempMR

    #Copy data into copy of array. I don't know if I really have to do this tho. -> We do, because we aren't originally provided with a subarray. We are only provided with the big array and indices. 
    #Also, I don't know if I necessarily have to loop through numbers. I would like to just go through elements in the original array. -> I have to again remember that len_temp is just a value and we cannot loop through it so we can only use range. We are not provided with sublist. 
    
    for i in range(len_tempLM):
        tempLM[i] = arr[l+i]

    for i in range(len_tempMR):
        tempMR[i] = arr[m+1+i]
    


    #initial index of subarray1, acts like a pointer. 
    i = 0
    #Initial index of subarray2
    j = 0
    #initial index of merged array
    k = l

    #We just have to go through the messy combinations. 

    #Scenario where we have to compare items from both arrays. 
    #We cannot make this <= because len_temp represents length of array while i and j represent indices. len_temp will always be 1 larger than max index. This way, we know we went through everything. 
    while i < len_tempLM and j < len_tempMR:
        if tempLM[i] <= tempMR[j]:
            #Modifying the original array
            #We start changing from the beginning of the array starting at index k
            arr[k] = tempLM[i]
            i += 1
        else:
            arr[k] = tempMR[j]
            j += 1
        k += 1

    while i < len_tempLM:
        #Nothing to compare with
        arr[k] = tempLM[i]
        i += 1
        k += 1

    while j < len_tempMR:
        arr[k] = tempMR[j]
        j += 1
        k += 1
    



def mergesort(arr, l, r):
    if l < r:
        mid = (l+r) // 2
        
        #*We are still always passing the original array into the merge method. The merge method just keeps making small changes. 

        mergesort(arr, l, mid)
        mergesort(arr, mid+1, r)
        merge(arr, l, mid, r)

arr = [4, 1, 3, 7, 3, 5]
mergesort(arr, 0, len(arr)-1)
print(arr)
```
